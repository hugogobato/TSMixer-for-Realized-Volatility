# F-tests

```python
import numpy as np
import scipy

def f_test(x, y):
    x = np.array(x)
    y = np.array(y)
    variance1=np.var(x, ddof=1)
    variance2=np.var(y, ddof=1)
    # Ensuring the larger variance is the numerator
    if variance1 > variance2:
      f_value = variance1 / variance2
      df1 = len(x) - 1  # degrees of freedom for x
      df2 = len(y) - 1  # degrees of freedom for y
    else:
      f_value = variance2 / variance1
      df1 = len(y) - 1  # degrees of freedom for y
      df2 = len(x) - 1  # degrees of freedom for x
    p = 1-scipy.stats.f.cdf(f_value, df1, df2) #find p-value of F test statistic
    return f_value, p

models = [TimesNet, NHITS,NBEATSx,TFT, LSTM]
models_names = ["TimesNet","NHITS","NBEATSx","TFT", "LSTM"]
for j in range(len(models)):
  print(f"{models_names[j]}")
  for i in range(len(NBEATSx.columns)):
    print(f"{models[j].columns[i]}: {f_test(models[j][models[j].columns[i]], TSMixer[models[j].columns[i]])}")
```
